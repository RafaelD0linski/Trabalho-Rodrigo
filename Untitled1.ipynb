import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

print("Bibliotecas importadas com sucesso!")


iris = load_iris()
X = iris.data
y = iris.target

print("Shape dos dados:", X.shape)
print("Classes reais:", iris.target_names)
print("\nPrimeiras linhas do dataset:")
pd.DataFrame(X, columns=iris.feature_names).head()


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Dados normalizados!")


kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X_scaled)

print("Modelo K-means treinado com sucesso!")
print("Centr√≥ides dos clusters:\n", kmeans.cluster_centers_)


labels = kmeans.labels_
silhouette = silhouette_score(X_scaled, labels)

print("Silhouette Score:", silhouette)
print("\nDistribui√ß√£o dos clusters:")
unique, counts = np.unique(labels, return_counts=True)
for cluster, count in zip(unique, counts):
    print(f"Cluster {cluster}: {count} pontos")


plt.figure(figsize=(10,4))

# Plot 1: Clusters encontrados pelo K-means
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='X', s=200, label='Centr√≥ides')
plt.title("Clusters K-means")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.legend()

# Plot 2: Classes reais (para compara√ß√£o)
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=50)
plt.title("Classes Reais")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

plt.tight_layout()
plt.show()


```python
inertias = []
K_range = range(1, 11)

for k in K_range:
    kmeans_temp = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_temp.fit(X_scaled)
    inertias.append(kmeans_temp.inertia_)

plt.figure(figsize=(8,4))
plt.plot(K_range, inertias, marker='o', linewidth=2)
plt.title("M√©todo do Cotovelo (Elbow Method)")
plt.xlabel("N√∫mero de Clusters (K)")
plt.ylabel("In√©rcia (Soma das dist√¢ncias)")
plt.grid(True)
plt.show()

print("üí° O 'cotovelo' indica o n√∫mero ideal de clusters!")
```
